```python
import re
import string # Added string import for punctuation removal

def clean_text(text):
 # Step 1: Convert to lowercase for uniformity
 text = text.lower()
 # Step 2: Remove email addresses
 text = re.sub(r'\S+@\S+', '', text)
 # Step 3: Remove standalone numbers
 text = re.sub(r'\b\d+\b', '', text)
 # Step 4: Remove hashtags
 text = re.sub(r'#\w+', '', text)
 # Step 5: Remove punctuation
 text = text.translate(str.maketrans('', '', string.punctuation))
 # Step 6: Remove extra whitespace
 text = re.sub(r'\s+', ' ', text).strip()
 return text

# Initialize raw_texts using the 'text' variable from the previous cell
# If 'text' is not defined in the current session, it will use the default value.
# To ensure independence, it's best to explicitly define it here if needed.
raw_texts = ["omg!!! I LOOOVE this movie Dhurandhar 2 sooo much!!! best everrrr!!! This product is terrible... waste of money!!! #disappointed hey u there??? reply asap!!! I can't believe this happened... like seriously??? LOL that was funny af!!! worst service ever!!!!! never coming back :( 100% satisfied with the purchase will buy again!!! wHaT iS gOiNg On here??? This is gr8!!! totally worth it ughhh I hate waiting in long lines hahaha can't stop laughing!!! The food was okaayyy... not that great tbh pls send the details ASAP!!! idk what to do anymore BEST. DAY. EVER!!! why is this sooo slowwww??? Thx for the help!!! appreciate it"]

cleaned_texts = [clean_text(t) for t in raw_texts]
print("Before vs After Cleaning:")
print("=" * 70)
for i in range(len(raw_texts)):
    print(f"\n Example {i+1}:")
    print(f"  BEFORE: {raw_texts[i]}")
    print(f"  AFTER : {cleaned_texts[i]}")
```

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize, sent_tokenize

def tokenize_text(text, mode='word'):
    """
    Tokenize text into words or sentences.

    Parameters:
        text (str): Input text
        mode (str): 'word' for word tokens, 'sentence' for sentence tokens

    Returns:
        list: List of tokens
    """
    if mode == 'word':
        return word_tokenize(text)          # Split into individual words
    elif mode == 'sentence':
        return sent_tokenize(text)          # Split into sentences
    else:
        raise ValueError("mode must be 'word' or 'sentence'")


# Demonstrate word tokenization
sample_text = raw_texts[0]  # Use first text for demo
word_tokens = tokenize_text(sample_text, mode='word')
sentence_tokens = tokenize_text(sample_text, mode='sentence')

print(" Original Text:")
print(f"  {sample_text}")

print(f"\n Word Tokens ({len(word_tokens)} tokens):")
print(f"  {word_tokens}")

print(f"\n Sentence Tokens ({len(sentence_tokens)} sentences):")
for s in sentence_tokens:
    print(f"  → {s}")

# Apply word tokenization to all cleaned texts
tokenized_texts = [tokenize_text(text, mode='word') for text in cleaned_texts]
print(f"\n Tokenization complete for all {len(tokenized_texts)} documents.")
```

```python
import nltk
nltk.download('stopwords') # Ensure stopwords are downloaded
from nltk.corpus import stopwords # Import stopwords module

# Load English stopwords from NLTK
stop_words = set(stopwords.words('english'))

# Add custom domain-specific stopwords
custom_stopwords = {'also', 'would', 'could', 'said', 'like', 'one', 'two'}
stop_words.update(custom_stopwords)

print(f" Total stopwords loaded: {len(stop_words)}")
print(f"   Sample: {sorted(list(stop_words))[:20]}")


def remove_stopwords(tokens):
    """
    Remove stopwords from a list of tokens.
    Also removes very short tokens (length <= 1).

    Parameters:
        tokens (list): List of word tokens

    Returns:
        list: Filtered token list without stopwords
    """
    return [
        token for token in tokens
        if token not in stop_words        # Not a stopword
        and len(token) > 1               # Not a single character
        and token.isalpha()              # Only alphabetic tokens
    ]


# Apply stopword removal
filtered_texts = [remove_stopwords(tokens) for tokens in tokenized_texts]
```

```python
from nltk.stem import PorterStemmer
from nltk.stem.snowball import SnowballStemmer # Import SnowballStemmer

#Initialize stemmers
porter_stemmer = PorterStemmer()       # Most common, aggressive
snowball_stemmer = SnowballStemmer('english')  # Improved version of Porter


def apply_stemming(tokens, stemmer='porter'):
    """
    Apply stemming to a list of tokens.

    Parameters:
        tokens (list): List of word tokens
        stemmer (str): 'porter' or 'snowball'

    Returns:
        list: Stemmed tokens
    """
    if stemmer == 'porter':
        return [porter_stemmer.stem(token) for token in tokens]
    elif stemmer == 'snowball':
        return [snowball_stemmer.stem(token) for token in tokens]


# Demo: Compare Porter vs Snowball on same words
demo_words = ['running', 'happily', 'studies', 'information', 'processing',
              'beautiful', 'generously', 'languages', 'prediction', 'classification']

print(" Stemming Comparison: Porter vs Snowball")
print("-" * 55)
print(f"{'Original':<20} {'Porter':<17} {'Snowball':<15}")
print("-" * 55)
for word in demo_words:
    porter = porter_stemmer.stem(word)
    snowball = snowball_stemmer.stem(word)
    print(f"{word:<20} {porter:<17} {snowball:<15}")

# Apply stemming to our filtered texts
stemmed_texts = [apply_stemming(tokens, stemmer='snowball') for tokens in filtered_texts]
print(f"\n Stemming applied to all documents.")
print(f" Sample: {stemmed_texts[0][:8]}")
```

```python
# Join tokens back into strings for vectorizer input
processed_corpus = [' '.join(tokens) for tokens in stemmed_texts]

print(" Processed Corpus (stemmed, stopwords removed):")
for i, doc in enumerate(processed_corpus, 1):
    print(f"  Doc {i}: {doc[:70]}...")

# ─── TF-IDF Vectorization ─────────────────────────────────────────────────────
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

tfidf_vectorizer = TfidfVectorizer(
    max_features=30,      # Keep top 30 most important features
    ngram_range=(1, 2),   # Include unigrams and bigrams
    min_df=1,             # Word must appear in at least 1 document
    sublinear_tf=True     # Apply log normalization to TF
)

# Fit and transform the corpus
tfidf_matrix = tfidf_vectorizer.fit_transform(processed_corpus)
feature_names = tfidf_vectorizer.get_feature_names_out()

print(f"\n TF-IDF Matrix Shape: {tfidf_matrix.shape}")
print(f"   → {tfidf_matrix.shape[0]} documents × {tfidf_matrix.shape[1]} features")

# Convert to DataFrame for visualization
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=feature_names,
    index=[f'Doc_{i+1}' for i in range(len(processed_corpus))]
)

print("\n TF-IDF Matrix (top 15 features):")
print(tfidf_df.iloc[:, :15].round(3).to_string())

# Show top TF-IDF words per document
print("\n Top 5 Keywords Per Document (by TF-IDF score):")
print("-" * 60)
for i, doc_name in enumerate(tfidf_df.index):
    top_words = tfidf_df.iloc[i].nlargest(5)
    words_str = ', '.join([f"{w}({v:.2f})" for w, v in top_words.items() if v > 0])
    print(f"  {doc_name}: {words_str}")
```
