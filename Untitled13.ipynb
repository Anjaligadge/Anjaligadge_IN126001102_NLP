```python
from datasets import load_dataset
dataset = load_dataset("conll2003")
```

assigning labals

```python
# getting POS tag labels (like noun, verb, adjective)
label_list = dataset["train"].features["pos_tags"].feature.names

# converting id to label (0 → NN, )
id2label = {i: label for i, label in enumerate(label_list)}

# converting label to id (NN → 0,)
label2id = {label: i for i, label in enumerate(label_list)}
```

Loading  tokenizer

```python
#DistilBERT is a smaller, faster, and lighter version of BERT so becz i using cpu for training
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
```

Tokenization + Label Alignment

```python
# this function converts text into tokens and aligns labels correctly
def tokenize_and_align_labels(examples):

    # tokenizing input words
    tokenized_inputs = tokenizer(
        examples["tokens"],   # list of words
        truncation=True,      # cut long sentences
        is_split_into_words=True  # words already split
    )

    labels = [] #here we will store aligned labels

    # looping through each sentence
    for i, label in enumerate(examples["pos_tags"]):

        word_ids = tokenized_inputs.word_ids(batch_index=i)
        previous_word_idx = None
        label_ids = []

        # aligning labels with tokens
        for word_idx in word_ids:

            if word_idx is None:
                # special tokens like [CLS], [SEP]
                label_ids.append(-100)

            elif word_idx != previous_word_idx:
                # first token of word → keep label
                label_ids.append(label[word_idx])

            else:
                # subword tokens → ignore
                label_ids.append(-100)

            previous_word_idx = word_idx

        labels.append(label_ids)

    # adding labels to tokenized data
    tokenized_inputs["labels"] = labels
    return tokenized_inputs
```

Apply preprocessing

```python
# applying tokenization to full dataset
tokenized_datasets = dataset.map(
    tokenize_and_align_labels,
    batched=True
)
```

```python
# loading pre-trained model for token classification
from transformers import AutoModelForTokenClassification

model = AutoModelForTokenClassification.from_pretrained(
    "distilbert-base-uncased",
    num_labels=len(label_list),   # number of POS tags
    id2label=id2label,
    label2id=label2id
)
```

Training settings

```python
from transformers import TrainingArguments

training_arguments = TrainingArguments(
    output_dir="./results",   # where model will be saved

    learning_rate=2e-5,       # how fast model learns
    per_device_train_batch_size=8,  # batch size becouse we using cpu for traing so
    per_device_eval_batch_size=8,

    num_train_epochs=2,       # how many times model sees data.its

    weight_decay=0.01,        # prevents overfitting

    logging_dir="./logs",     # logs folder
    logging_steps=10# log progress after every 10 steps
)
```

Evaluation metric
to **check** How correct our  model predictions are compared to actual labels

```python
import numpy as np
import evaluate # Import the evaluate library


metric = evaluate.load("seqeval") # Use evaluate.load instead of datasets.load_metric

def compute_metrics(p):
    # Access predictions and label_ids from the PredictionOutput object
    predictions = p.predictions
    labels = p.label_ids

    # taking highest probability label
    predictions = np.argmax(predictions, axis=2)

    # removing ignored labels (-100)
    true_predictions = [
        [label_list[p] for (p, l) in zip(pred, lab) if l != -100]#-100 = special tokens, subwords
        for pred, lab in zip(predictions, labels)
    ]

    true_labels = [
        [label_list[l] for (p, l) in zip(pred, lab) if l != -100]
        for pred, lab in zip(predictions, labels)
    ]

    # calculating precision, recall, f1
    results = metric.compute(
        predictions=true_predictions,
        references=true_labels
    )

    return {
        "precision": results["overall_precision"],
        "recall": results["overall_recall"],
        "f1": results["overall_f1"],
    }
```

# using smaller subset of data to reduce training time on CPU

```python
from transformers import Trainer, DataCollatorForTokenClassification   #importing the training utility

data_collator = DataCollatorForTokenClassification(tokenizer=tokenizer)

trainer = Trainer(
    model=model,
    args=training_arguments,

    # using small dataset (faster for CPU)
    train_dataset=tokenized_datasets["train"].select(range(2000)),
    eval_dataset=tokenized_datasets["validation"].select(range(500)),# using 500 samples to test model

    data_collator=data_collator,
    compute_metrics=compute_metrics
)
```

Train model

```python
# training starts here
trainer.train()
```

Evaluate model

```python
predictions_output = trainer.predict(tokenized_datasets["validation"].select(range(500)))
metrics = compute_metrics(predictions_output)
print(metrics)
```

Inference


```python
# testing model on custom sentence
sentence = "my friends are cool but they  dont know how dance ediots"

from transformers import AutoTokenizer, AutoModelForTokenClassification

tokens = tokenizer(sentence, return_tensors="pt")

outputs = model(**tokens).logits
predictions = outputs.argmax(dim=2)

# converting ids back to labels
predicted_labels = [id2label[p.item()] for p in predictions[0]]

# printing token + label
print(list(zip(tokenizer.convert_ids_to_tokens(tokens["input_ids"][0]), predicted_labels)))
```
