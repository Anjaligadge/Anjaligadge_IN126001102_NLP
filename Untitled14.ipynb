chat prompt temple  to build prompt with 2 diffrent role , system, human

```python
# role definitions — each role gives the ai a different personality/behaviour
role_descriptions = {
    "teacher":     "you are a patient and clear teacher. explain concepts step by step with simple examples.",
    "interviewer": "you are a senior technical interviewer. ask 3 challenging but fair questions about the topic.",
    "motivator":   "you are an enthusiastic motivator. make the topic exciting and inspire the learner to dive deeper."
}

# Build a ChatPromptTemplate using a list of (role, message) tuples
# {system_message} and {topic} are placeholders filled at runtime
chat_template = ChatPromptTemplate.from_messages([
    ("system", "{system_message}"),   # Controls AI behaviour
    ("human",  "Teach me about {topic}")  # The user's request
])

print("=" * 55)
print("TASK 4 OUTPUT")
print("=" * 55)

# test all three roles with the same topic
topic = "Neural Networks"

for role, description in role_descriptions.items():
    # format_messages() returns a list of message objects (SystemMessage, HumanMessage)
    messages = chat_template.format_messages(
        system_message=description,
        topic=topic
    )
    print(f"\nROLE: {role.upper()}")
    for msg in messages:
        # msg.type gives us 'system' or 'human'; msg.content gives the text
        print(f"  [{msg.type.upper()}]: {msg.content}")
```

input validation layer

validate user input before  passing them into any template

```python
# define allowed values as constants at the to
VALID_AUDIENCES = ["beginner", "intermediate", "expert"]
VALID_TONES     = ["formal", "casual", "fun"]

# approach a: raise an error on invalid input (strict validation)
def validate_inputs(audience: str, tone: str) -> tuple:
    """
    Validate that audience and tone are from the allowed lists.
    Raises ValueError if either is invalid.
    Returns (audience, tone) if both are valid.
    """
    if audience not in VALID_AUDIENCES:
        raise ValueError(
            f"Invalid audience '{audience}'. "
            f"Allowed values: {VALID_AUDIENCES}"
        )
    if tone not in VALID_TONES:
        raise ValueError(
            f"Invalid tone '{tone}'. "
            f"Allowed values: {VALID_TONES}"
        )
    return audience, tone


#  approach b: Assign defaults on invalid input (lenient validation)
def validate_with_defaults(audience: str, tone: str,
                           default_audience: str = "beginner",
                           default_tone: str = "casual") -> tuple:
    """
    Validate inputs and silently fall back to defaults if invalid.
    Returns (audience, tone) — always valid.
    """
    if audience not in VALID_AUDIENCES:
        print(f"   Invalid audience '{audience}' → using default '{default_audience}'")
        audience = default_audience
    if tone not in VALID_TONES:
        print(f"   Invalid tone '{tone}' → using default '{default_tone}'")
        tone = default_tone
    return audience, tone


print("=" * 55)
print("TASK 5 OUTPUT")
print("=" * 55)

# Test 1: Valid inputs — should pass without error
print("\nTest 1 — Valid inputs:")
a, t = validate_inputs("beginner", "fun")
print(f"   Passed: audience='{a}', tone='{t}'")

# Test 2: Invalid audience — should raise ValueError
print("\nTest 2 — Invalid audience (strict):")
try:
    validate_inputs("alien", "fun")
except ValueError as e:
    print(f"   Caught error: {e}")

# Test 3: Invalid tone — should raise ValueError
print("\nTest 3 — Invalid tone (strict):")
try:
    validate_inputs("beginner", "robotic")
except ValueError as e:
    print(f"   Caught error: {e}")

# Test 4: Using the lenient approach with defaults
print("\nTest 4 — Invalid inputs (lenient, uses defaults):")
a, t = validate_with_defaults("martian", "robotic")
print(f"   Resolved: audience='{a}', tone='{t}'")
```

prompt generater app

build single  genertor_prompt() that combine all validation , style selecting, formating

```python

# Keeping templates in a dict makes adding new styles trivial
style_templates = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} teaching style, step by step"
    ),
    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Ask 3 {tone} interview questions about {topic} suitable for {audience}"
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} as a {tone} story that a {audience} would enjoy"
    ),
}

VALID_STYLES = list(style_templates.keys())


def generate_prompt(topic: str, audience: str, tone: str, style: str) -> str:
    """
    Generate a structured prompt from user inputs.

    Parameters:
        topic    (str) : The subject to explain.
        audience (str) : Target audience — must be in VALID_AUDIENCES.
        tone     (str) : Communication tone — must be in VALID_TONES.
        style    (str) : Prompt style — one of 'teaching', 'interview', 'storytelling'.

    Returns:
        str: The formatted prompt string, ready to send to an LLM.
    """
    # step 1 — Validate audience and tone (raises ValueError if invalid)
    audience, tone = validate_inputs(audience, tone)

    # step 2 — Validate style
    if style not in style_templates:
        raise ValueError(
            f"Invalid style '{style}'. Allowed styles: {VALID_STYLES}"
        )

    # step 3 — Select the correct template and format it
    selected_template = style_templates[style]
    return selected_template.format(topic=topic, audience=audience, tone=tone)


print("=" * 55)
print("TASK 6 OUTPUT")
print("=" * 55)

# demonstrate all three styles
test_runs = [
    ("singing", "beginner",     "fun",      "storytelling"),
    ("singing", "intermediate", "casual",   "teaching"),
    ("singing", "expert",       "formal",   "interview"),
]

for topic, audience, tone, style in test_runs:
    result = generate_prompt(topic, audience, tone, style)
    print(f"\nStyle [{style.upper()}]:")
    print(f"  {result}")

# Show that validation catches bad input
print("\n--- Error handling demo ---")
try:
    generate_prompt("AI", "alien", "fun", "teaching")
except ValueError as e:
    print(f" {e}")
```

```python
# one template — defined exactly once
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# five different input sets — diverse topics, audiences, and tones
five_inputs = [
    {"topic": "Artificial Intelligence", "audience": "beginners",   "tone": "friendly"},
    {"topic": "Python Programming",      "audience": "kids",         "tone": "fun"},
    {"topic": "Deep Learning",           "audience": "engineers",    "tone": "technical"},
    {"topic": "Blockchain",              "audience": "investors",    "tone": "formal"},
    {"topic": "Quantum Computing",       "audience": "researchers",  "tone": "casual"},
]

print("=" * 60)
print("TASK 7 OUTPUT — ONE Template, FIVE Different Outputs")
print("=" * 60)
print(f"Template: '{reusable_template.template}'")
print()

for i, inputs in enumerate(five_inputs, start=1):
    # ** unpacks the dict: format(topic=..., audience=..., tone=...)
    result = reusable_template.format(**inputs)
    print(f"Input {i}: {result}")

print()
print(" Same template structure → 5 unique, correctly formatted prompts.")
```

full pipeline demo, user input-validation-prompt templeate-dynamic prompt- output

```python
def run_full_pipeline(topic: str, audience: str, tone: str, style: str) -> None:
    """
    Demonstrate the complete prompt pipeline:
    User Input → Validation → Template → Output
    """
    print("=" * 55)
    print("FULL PIPELINE RUN")
    print("=" * 55)
    print(f"Input  → topic='{topic}', audience='{audience}', tone='{tone}', style='{style}'")

    try:
        # Step 1: Validate
        aud, tn = validate_inputs(audience, tone)
        print(f"Validate →  audience='{aud}', tone='{tn}'")

        # Step 2: Generate prompt using the generate_prompt app
        final_prompt = generate_prompt(topic, aud, tn, style)
        print(f"Template → '{style_templates[style].template}'")
        print(f"Output   → {final_prompt}")

    except ValueError as e:
        print(f"Validate →  Error: {e}")


# run the full pipeline
run_full_pipeline(
    topic    = "singing",
    audience = "beginner",
    tone     = "fun",
    style    = "storytelling"
)

print()

# run with invalid input to see error handling
run_full_pipeline(
    topic    = "singing",
    audience = "alien",  # invalid
    tone     = "fun",
    style    = "teaching"
)
```
