```python
# Import required libraries
import pandas as pd
import numpy as np
import torch

# Sklearn for splitting and evaluation
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Hugging Face transformers
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from torch.optim import AdamW

# PyTorch dataset and dataloader
from torch.utils.data import Dataset, DataLoader, Subset
```

```python
df = pd.read_csv("/content/IMDB Dataset.csv", sep=',', engine='python', on_bad_lines='skip')

# Convert labels to numeric (positive=1, negative=0)
df['sentiment'] = df['sentiment'].str.strip().str.lower()
df['sentiment'] = df['sentiment'].map({'positive': 1, 'negative': 0})
df = df.dropna(subset=['sentiment'])
df['sentiment'] = df['sentiment'].astype(int)
```

```python
# Split data into train (80%) and temp (20%)
train_texts, temp_texts, train_labels, temp_labels = train_test_split(
    df['review'], df['sentiment'], test_size=0.2, random_state=42)

# Split temp into validation and test (10% each)
val_texts, test_texts, val_labels, test_labels = train_test_split(
    temp_texts, temp_labels, test_size=0.5, random_state=42)
```

```python
# Load BERT tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Function to tokenize text
def tokenize(texts):
    return tokenizer(
        texts.tolist(),           # Convert pandas series to list
        padding=True,             # Pad sequences to same length
        truncation=True,          # Truncate long sentences
        max_length=128,           # Max token length
        return_tensors="pt"       # Return PyTorch tensors
    )

# Apply tokenization
train_encodings = tokenize(train_texts)
val_encodings = tokenize(val_texts)
test_encodings = tokenize(test_texts)
```

```python
# Create custom dataset class for PyTorch
class SentimentDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels.tolist()

    def __getitem__(self, idx):
        # Get input_ids, attention_mask and label
        item = {key: val[idx] for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)
```

```python
# Create custom dataset class for PyTorch
class SentimentDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels.tolist()

    def __getitem__(self, idx):
        # Get input_ids, attention_mask and label
        item = {key: val[idx] for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)
```

```python
# Create dataset objects
train_dataset = SentimentDataset(train_encodings, train_labels)
val_dataset = SentimentDataset(val_encodings, val_labels)
test_dataset = SentimentDataset(test_encodings, test_labels)

# Reduce dataset size to avoid long training time (important for BERT)
train_dataset = Subset(train_dataset, range(min(500, len(train_dataset))))
val_dataset = Subset(val_dataset, range(min(200, len(val_dataset))))
```

```python
# Create dataloaders for batching
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8)
test_loader = DataLoader(test_dataset, batch_size=8)
```

```python
# Check for GPU availability
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load pre-trained BERT model for classification
model = AutoModelForSequenceClassification.from_pretrained(
    "bert-base-uncased", num_labels=2)

# Move model to device (GPU/CPU)
model.to(device)
```

```python
# Freeze all BERT layers (only classifier will train)
for param in model.bert.parameters():
    param.requires_grad = False
```

```python
# Define optimizer
optimizer = AdamW(model.parameters(), lr=2e-5)
```

```python
# Function to train model for one epoch
def train_one_epoch(model, loader):
    model.train()
    total_loss = 0

    for batch in loader:
        optimizer.zero_grad()

        # Move data to device
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        # Forward pass
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss

        # Backward pass
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    return total_loss / len(loader)
```

```python
# Function to evaluate model
def evaluate(model, loader):
    model.eval()
    preds, true_labels = [], []

    with torch.no_grad():
        for batch in loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            # Forward pass
            outputs = model(input_ids, attention_mask=attention_mask)
            logits = outputs.logits

            # Get predictions
            preds.extend(torch.argmax(logits, dim=1).cpu().numpy())
            true_labels.extend(labels.cpu().numpy())

    return preds, true_labels
```

```python
# Number of epochs (keep small for faster execution)
EPOCHS = 2

for epoch in range(EPOCHS):
    print(f"\nEpoch {epoch+1}")

    # Train model
    train_loss = train_one_epoch(model, train_loader)

    # Evaluate model
    preds, labels = evaluate(model, val_loader)

    # Calculate accuracy
    acc = accuracy_score(labels, preds)

    print(f"Train Loss: {train_loss:.4f}")
    print(f"Validation Accuracy: {acc:.4f}")
```

```python
# Evaluate on test data
preds, labels = evaluate(model, test_loader)

print("Accuracy:", accuracy_score(labels, preds))
print("Precision:", precision_score(labels, preds))
print("Recall:", recall_score(labels, preds))
print("F1 Score:", f1_score(labels, preds))

# Confusion Matrix
print("Confusion Matrix:\n", confusion_matrix(labels, preds))
```
